```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
%matplotlib inline
```

```python
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
circut = QuantumCircuit(qr, cr)
# Below added later. Hadamand
circut.h(qr)
circut.measure(qr, cr)
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circut, backend= simulator, shots= 102400).result()
plot_histogram(result.get_counts())
```
